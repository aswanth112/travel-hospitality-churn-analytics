```python
# Week 1 - Cell 2: Check Missing Values and Duplicates
print("=" * 50)
print("MISSING VALUES PER COLUMN:")
print("=" * 50)
missing = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage': missing_percent
})
print(missing_df[missing_df['Missing Count'] > 0])

print("\n" + "=" * 50)
print("DUPLICATE ROWS:")
print("=" * 50)
print("Total duplicate rows:", df.duplicated().sum())

print("\n" + "=" * 50)
print("BASIC STATISTICS:")
print("=" * 50)
print(df.describe())
```

```python
# Week 1 - Cell 3: Data Cleaning

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)
print(f" Duplicates removed. New shape: {df.shape}")

# 2. Drop 'company' column (94% missing - not useful)
df.drop(columns=['company'], inplace=True)
print(" Dropped 'company' column")

# 3. Fill missing values
df['agent'].fillna(0, inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['children'].fillna(0, inplace=True)
print(" Filled missing values in agent, country, children")

# 4. Fix negative ADR values (set to 0)
df['adr'] = df['adr'].apply(lambda x: 0 if x < 0 else x)
print(" Fixed negative ADR values")

# 5. Cap extreme ADR outliers (above 3500 are errors)
df['adr'] = df['adr'].clip(upper=3500)
print(" Capped ADR outliers")

# 6. Convert arrival_date_month to number for sorting
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['arrival_date_month_num'] = df['arrival_date_month'].map(month_map)
print(" Converted month names to numbers")

# 7. Create full arrival date column
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month_num'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str)
)
print(" Created arrival_date column")

# 8. Final check
print(f"\n Final dataset shape: {df.shape}")
print(f" Missing values remaining:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
```

```python
# Week 1 - Cell 4: Save Cleaned Dataset

import os

# Create processed folder if it doesn't exist
os.makedirs(r"C:\Users\aswan\OneDrive\Desktop\Infotact internship 2\data\processed", exist_ok=True)

# Save cleaned data
df.to_csv(r"C:\Users\aswan\OneDrive\Desktop\Infotact internship 2\data\processed\hotel_bookings_cleaned.csv", index=False)

print(" Cleaned dataset saved successfully!")
print(f" Final shape: {df.shape}")
print(f" Columns: {df.columns.tolist()}")
```

```python
# checking cancellation rate in the dataset
total_bookings = len(df)
cancelled = df['is_canceled'].sum()
not_cancelled = total_bookings - cancelled

print("Total bookings:", total_bookings)
print("Cancelled bookings:", cancelled)
print("Not cancelled bookings:", not_cancelled)
print("Cancellation rate:", round((cancelled / total_bookings) * 100, 2), "%")
```

```python
# checking cancellation rate by hotel type
hotel_cancel = df.groupby('hotel')['is_canceled'].mean() * 100

print("Cancellation rate by hotel type:")
print(hotel_cancel.round(2))
```

```python
# checking monthly booking trends
monthly_bookings = df.groupby('arrival_date_month')['is_canceled'].count()

print("Total bookings per month:")
print(monthly_bookings)
```

```python
# checking average daily rate by hotel type
adr_by_hotel = df.groupby('hotel')['adr'].mean()

print("Average Daily Rate by Hotel Type:")
print(adr_by_hotel.round(2))
```

```python
# checking lead time vs cancellation
lead_time_cancel = df.groupby('is_canceled')['lead_time'].mean()

print("Average lead time by cancellation status:")
print(lead_time_cancel.round(2))
print("\n0 = Not Cancelled")
print("1 = Cancelled")
```

```python
# checking cancellation rate by market segment
segment_cancel = df.groupby('market_segment')['is_canceled'].mean() * 100

print("Cancellation rate by market segment:")
print(segment_cancel.round(2))
```
